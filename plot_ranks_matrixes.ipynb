```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


def visualize(data):
# Создаем тепловую карту
    plt.figure(figsize=(10, 6))  # Задаем размер графика
    sns.heatmap(data, cmap='viridis', cbar_kws={'label': 'Значения'})  # 'viridis' — цветовая карта
    plt.title('Тепловая карта')
    plt.xlabel('Регионы')
    plt.ylabel('Стимулы')
    plt.show()
```

```python
auc_train_matrix = './ranks_matrix/HC_raw_matrix_auc.npy'
auc_test_matrix = './ranks_matrix/test_matrix_auc.npy'

max_train_matrix = './ranks_matrix/HC_raw_matrix_max.npy'
max_test_matrix = './ranks_matrix/test_matrix_max.npy'

min_train_matrix = './ranks_matrix/HC_raw_matrix_min.npy'
min_test_matrix = './ranks_matrix/test_matrix_min.npy'
```

```python
matrix = np.load(auc_train_matrix)
N = matrix.shape[0]  # Длина массива
sub_num = N // 5

subjects = np.array_split(matrix, sub_num)
for idx, sub in enumerate(subjects):
    print(f'sub-{idx:02d}')
    visualize(subjects[idx])
```

```python
matrix = np.load(auc_test_matrix)
N = matrix.shape[0]  # Длина массива
sub_num = N // 5

subjects = np.array_split(matrix, sub_num)
for idx, sub in enumerate(subjects):
    print(f'sub-{idx:02d}')
    visualize(subjects[idx])
```
